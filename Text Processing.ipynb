```python
'''
    Makes a list of all the words and their frequency of occurance in the text
of Shakespeare's Hamlet and then identifies the most commonly used words in a 
pie chart.  Then prodeces plot to see if Zipf's law is present.
'''
```

```python
%pylab inline
```

```python
#opens text file in this case Hamlet
f = open(r"C:\Users\Evan\Documents\Python Projects\hamlet.txt")
text = f.read()
#print text
```

```python
text = text.lower() #makes everything lowercase
```

```python
#removes punctuation
text = text.replace(':','')
text = text.replace(',','')
text = text.replace('!','')
text = text.replace('?','')
text = text.replace('[','')
text = text.replace(']','')
text = text.replace('-','')
text = text.replace('*','')
text = text.replace(';','')
text = text.replace('.','')
```

```python
words = text.split()  #splits all words by whitespace
```

```python
#counts appearance of each word in Hamlet
wordcount ={}
for word in words:
    if word in wordcount:
        wordcount[word]+=1
    else:
        wordcount[word] = 1
#print wordcount
```

```python
#puts words and corrisponding appearance into colm vector
wordcount.items()
```

```python
#gets the number that each word occured
def get_count(x):
    return x[1]
```

```python
#arranges words based on frequency of occurance
sorted_words = sorted(wordcount.items(), key=get_count, reverse=True)
#print sorted_words
```

```python
counts = [get_count(pair) for pair in sorted_words]
labels = [pair[0] for pair in sorted_words]
```

```python
#Gives list of most used words and how many times it was used
cList = []
lList = []
for i in range(len(counts)):
    if counts[i] > 100:
        cList.append(counts[i])
        lList.append(labels[i])
print cList
print lList
```

```python
#pie chart of word usage
figure(figsize=(9,9))
p = pie(cList, labels=lList)
```

```python
figure(figsize=(7,7))
loglog(counts, 'r.')
xlim(0,3000)
ylim(0,1000)
xlabel('order of word occurance (most frequent to least frequent)')
ylabel('frequency of occurance')
```
