```python
'''
    Investment is based on risk and return.  To invest in a stock one wants to 
minimize the risk of losing money, but maximize the return on his or her 
investment.  These risk and return values come from the mean change in stock 
price on a daily.  Therefore, since stock price depends on the mean of its 
changing value, the stock prices can be represented as a normal distribution, 
where the mean of the normal distribution can indicate what returns are expected 
over a course of time. A stock price movement can be simulated using the 
formula: s(t+1) = (1 + μ + σε)*s(t).  In this equation, s(t+1) is the stock 
price for the next time t, μ represents the rate of growth per each iteration 
of t, σ represents the volatility or how much s changes each day, and ε 
represents the stochastic or random variable with a normal distribution.
'''
```

```python
#Plotting changing stock prices
st = 1
resultsList = []
tList = []
steps = 366
u = 0.0
sg = 0.05
for t in range(steps):
    e = random.normal(size = 5) #controls how many simulations are run
    st = (1 + u + sg*e)*st
    #print st
    resultsList.append(st)
    tList.append(t)
print st
#print mean(resultsList)
plot(tList, resultsList)
title('Stock price for 5 different trajectories over 365 days')
xlabel('change in number of days')
ylabel('stock price')
```

```python
#Plotting histogram of final prices
st = 1.
steps = 366
u = 0.0
sg = 0.05
for i in range(steps):
    e = random.normal(size = 10000) #controls how many simulations are run
    st = (1. + u + sg*e)*st
final = st[:-1]
hist(final, bins=50, normed = True, range = (0,5))
ylabel('Frequency of occurance')
xlabel('Final price of stock after 365 days')


#Analysing the distribution of final prices
expected = mean(final)
print 'expected stock value after 365 days:', expected

lose_money = sum(final < 1.) / 100
print "likelihood of losing money:", lose_money, "%"
print "most likely value of 365 day investment:",median(final)
```
